```python
import os
import json
```

```python
path = 'results_combine/'
files = os.listdir(path)
```

```python
jsons = []
for fn in files:
    with open(path + fn, 'r') as fnb:
        jsons.append(json.load(fnb))
```

```python
jsons[-3]
```

```python
table = []
for j in jsons:
    table.append({
        'id': 'cipta-' + j['config']['experiment_id'],
        'data': j['config']['data'],
        'rnn_cell': j['config']['model'],
        'optimizer': 'normal',
        'n_rnn_layers': j['config']['nlayers'],
        'dim_embed': j['config']['emsize'],
        'dim_hidden': j['config']['nhid'],
        'vocab_size': j['config'].get('vocab_size', 10000),
        'shuffling': '',
        'epochs': j['config']['epochs'],
        'last_valid_loss': j['epoch_logs'][-1]['val_loss'],
        'last_valid_ppl': j['epoch_logs'][-1]['val_ppl'],
        'test_loss': j['test_loss'],
        'test_ppl': j['test_ppl']
    })
```

```python
import pandas as pd
frame = pd.DataFrame(table).set_index('id').sort_index()
frame = frame[['data', 'rnn_cell', 'optimizer', 'n_rnn_layers', 'dim_embed', 'dim_hidden', 'vocab_size', 'shuffling',
               'epochs', 'last_valid_loss', 'last_valid_ppl', 'test_loss', 'test_ppl']]
frame[frame['epochs'] > 10]
```

```python
frame['rnn_cell'].unique()
```

```python
penn = frame[frame['data'].str.contains('penn')]
penn = penn[penn['epochs'] == 20]
penn = penn.drop(['data'], axis=1)
len(penn)
```

```python
gut = frame[frame['data'].str.contains('gutenberg')]
gut = gut[gut['epochs'] == 20]
gut = gut.drop(['data'], axis=1)
len(gut)
```

```python
penn.to_csv('cipta_penn.csv')
```

```python
gut.to_csv('cipta_gut.csv')
```
